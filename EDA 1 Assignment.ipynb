### 2.Statistical Summary:

```python
df.describe()
```

```python
df.median()
```

```python
# LB :
#Mean: 133 bpm with a standard deviation of 11.27, suggesting moderate variability.
# Range: The minimum value is quite low at 51.84 bpm, and the maximum reaches 214 bpm, which is significantly higher than typical fetal heart rates (normally between 110-160 bpm).
```

```python
# AC and UC:
# Both have very low mean values (~0.003 and 0.004, respectively) and high variability compared to their means.
# Negative values: Interestingly, there are negative values for AC and UC, which might indicate data entry errors or specific situations that need to be clarified. In reality, acceleration and contraction measures shouldn't be negative.
```

```python
# FM :
# It has a relatively low mean of 0.0099 but a high standard deviation of 0.0675, indicating a wide range of values.
# Range: It goes from -0.48 to 0.96, again showing negative values, which could indicate issues with the data.
```

```python
# ASTV :
# Mean: 47.01 with a high standard deviation of 18.81, showing significant variation in abnormal short-term variability.
# Range: This feature ranges from -63 to 162, with negative values being potentially problematic.
# The median value is 49, meaning half of the data has ASTV below this, but there are extreme cases with much higher variability.
```

```python
# NSP :
# Range: The variable ranges from -1.03 to 5, where negative values and those above 3 could represent non-standard classifications.
# Mean: The mean value of 1.30 suggests most cases might be normaL.
# Since this is likely a classification label, a closer look at its distribution and its relationship with other variables could provide more insights.
```

```python
# Tendency:
# Mean: 0.316, indicating the overall tendency might lean towards a lower value, but the range goes from -3 to 3.
# This wide range, including negative values, is intriguing and may warrant a closer look at what "tendency" signifies and its importance in interpreting the cardiotocographic data.
```

```python
#  Width:
# The range of values is quite large (from -174 to 357), with a standard deviation of 42.94. This feature may capture some important characteristics of the cardiotocographic readings, but it also appears to have extreme outliers or potential issues with data.
```
